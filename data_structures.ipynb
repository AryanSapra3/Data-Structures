## Section A: Data Types and Structures Questions

### 1. What are data structures, and why are they important?
A data structure is a way of organizing, storing, and managing data so it can be accessed and modified efficiently. They are important because the right structure (list, dictionary, set, tuple, tree, etc.) directly affects how fast and how simply a program can search, insert, delete, or iterate over data — good data structure choices lead to more efficient, readable, and scalable code.

### 2. Explain the difference between mutable and immutable data types with examples.
**Mutable** data types can be changed in place after creation — e.g. `list`, `dict`, and `set` (`my_list.append(4)` modifies the same object). **Immutable** data types cannot be changed once created — any modification produces a new object — e.g. `int`, `float`, `str`, and `tuple` (`"abc" + "d"` creates a brand-new string rather than altering the original).

### 3. What are the main differences between lists and tuples in Python?
- **Mutability:** lists are mutable (`[1, 2, 3]`), tuples are immutable (`(1, 2, 3)`).
- **Syntax:** lists use square brackets `[]`, tuples use parentheses `()`.
- **Performance:** tuples are generally faster and use less memory since they're fixed-size.
- **Use case:** lists suit collections that change; tuples suit fixed records and are hashable (usable as dictionary keys), which lists are not.

### 4. Describe how dictionaries store data.
Dictionaries store data as **key-value pairs**. Internally they use a hash table: each key is passed through a hash function to compute an index where the corresponding value is stored, which is why looking up a value by its key is very fast (average O(1)) regardless of the dictionary's size.

### 5. Why might you use a set instead of a list in Python?
Sets automatically enforce **uniqueness** (no duplicate elements) and offer much faster **membership testing** (`x in my_set` is on average O(1) vs O(n) for a list). Sets also provide built-in mathematical operations like union, intersection, and difference, which makes them ideal when you need distinct values or fast lookups rather than an ordered, indexable sequence.

### 6. What is a string in Python, and how is it different from a list?
A string is an **immutable** ordered sequence of characters, used to represent text. A list is a **mutable** ordered sequence that can hold elements of any type (not just characters). Because strings are immutable, operations like replacing a character always return a new string, whereas a list can be modified in place.

### 7. How do tuples ensure data integrity in Python?
Since tuples are immutable, once created, their contents cannot be accidentally (or intentionally) altered by any part of the program. This guarantees that data meant to stay constant — like fixed configuration values or coordinate pairs — remains reliable and predictable throughout the program's execution.

### 8. What is a hash table, and how does it relate to dictionaries in Python?
A hash table is a data structure that maps keys to values using a **hash function**, which converts a key into an index in an underlying array where the value is stored. Python's `dict` is implemented as a hash table, which is why key lookups, insertions, and deletions are all on average O(1).

### 9. Can lists contain different data types in Python?
Yes. Python lists are heterogeneous — a single list can hold a mix of types, e.g. `[1, "two", 3.0, [4, 5], True]`.

### 10. Explain why strings are immutable in Python.
Strings are immutable by design for several reasons: it makes them **hashable** so they can be used as dictionary keys or set members; it allows Python to safely **cache/reuse** string objects internally for memory efficiency; and it prevents unexpected side effects, since a string passed around the program can't be silently changed by another part of the code.

### 11. What advantages do dictionaries offer over lists for certain tasks?
Dictionaries provide direct, near-instant (O(1) average) access to a value via its key instead of scanning the whole collection (O(n) for a list). This makes them far better suited for lookup-heavy tasks — such as counting occurrences, caching results, or mapping identifiers to records — where a list would require slower linear searches.

### 12. Describe a scenario where using a tuple would be preferable over a list.
Representing a fixed record such as GPS coordinates `(latitude, longitude)` is a good case for a tuple: the two values always belong together and shouldn't change once set. Using a tuple communicates that intent, prevents accidental modification, and — since tuples are hashable — allows the coordinate pair to be used directly as a dictionary key or stored in a set.

### 13. How do sets handle duplicate values in Python?
Sets automatically discard duplicates — when you add a value that already exists in the set, nothing changes; each element is guaranteed to appear only once. This happens because sets are built on the same hashing mechanism as dictionaries.

### 14. How does the “in” keyword work differently for lists and dictionaries?
For a **list**, `in` checks whether a value exists among the list's elements by scanning sequentially (O(n)). For a **dictionary**, `in` checks whether a value exists among the dictionary's **keys** (not its values) using the hash table, making the check average O(1) — e.g. `"name" in my_dict` checks the keys, not `my_dict.values()`.

### 15. Can you modify the elements of a tuple? Explain why or why not.
No. Tuples are immutable, so once created, none of their elements can be reassigned, added, or removed. This is intentional — it lets tuples serve as safe, constant, hashable records. Note: if a tuple contains a mutable object (like a list), that inner object's *contents* can still be changed — but the tuple's own slots/references cannot be reassigned.

### 16. What is a nested dictionary, and give an example of its use case?
A nested dictionary is a dictionary whose values are themselves dictionaries, allowing hierarchical or structured data to be represented. Example use case: storing student records, where each student ID maps to a dictionary of their details:

```python
students = {
    "S101": {"name": "Alice", "age": 21, "grade": "A"},
    "S102": {"name": "Bob", "age": 22, "grade": "B"}
}
```

### 17. Describe the time complexity of accessing elements in a dictionary.
On average, accessing a value by its key in a dictionary is **O(1)** (constant time), thanks to the underlying hash table. In rare worst-case scenarios (many hash collisions), it can degrade to O(n), but Python's hashing implementation is designed to make this extremely unlikely in practice.

### 18. In what situations are lists preferred over dictionaries?
Lists are preferable when: data has a natural **order** that matters; you need to access elements by **position/index** rather than a meaningful key; the collection may contain **duplicate** values; or you simply need a straightforward, ordered sequence without the overhead of key-value pairing (e.g. a list of scores, a queue of tasks, a sequence of steps).

### 19. Why are dictionaries considered unordered, and how does that affect data retrieval?
Historically, dictionaries were considered unordered because their internal hash-table layout does not follow insertion order — the position of a key-value pair depends on the key's hash, not when it was added. (Note: since Python 3.7+, dictionaries do preserve insertion order as an implementation detail/guarantee, but conceptually they're still designed to be accessed by key rather than by position.) This means retrieval should always be done via keys (`my_dict["key"]`) rather than by assuming a particular positional order.

### 20. Explain the difference between a list and a dictionary in terms of data retrieval.
In a **list**, elements are retrieved by their **numeric index** (position), e.g. `my_list[0]`, and finding a specific value requires scanning the list (O(n)) unless you already know its index. In a **dictionary**, elements are retrieved by their **key**, e.g. `my_dict["name"]`, which is direct and fast (O(1) average) regardless of how many items the dictionary holds.
